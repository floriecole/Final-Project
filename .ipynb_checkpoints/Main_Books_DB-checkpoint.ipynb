```python
genres_df=pd.merge(books_df, goodreads_df, left_on='Title', right_on='book_title', how='inner')
genres_df.shape
```

```python
author_df = pd.read_csv('Resources/good_reads_final.csv')
author_df.head()
```

```python
author_df['book_title']=author_df['book_title'].replace('\n', '', regex=True)
author_df['author_name']=author_df['author_name'].replace('\n', '', regex=True)
author_df['birthplace']=author_df['birthplace'].replace('\n', '', regex=True)
author_df
```

```python
author_df.shape
```

```python
author_new_df.shape
```

```python
books_df_2.shape
```

```python
nationality_df=pd.merge(books_df_2, author_new_df, left_on='Author', right_on='author_name', how='inner')
nationality_df.shape
```
