```python
#Dependencies
import pandas as pd
import numpy as np
import pymysql
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float
pymysql.install_as_MySQLdb() 
```

```python
#Read the csv file into a pandas DataFrame
ratings_df = pd.read_csv('Resources/BX-Book-Ratings.csv',encoding="ISO-8859-1")
ratings_df.head()
```

```python
#Read the csv file into a pandas DataFrame
users_df = pd.read_csv('Resources/BX-Users.csv',encoding="ISO-8859-1")
users_df.head()
```

```python
users_df['Country'].replace('', np.nan, inplace=True)
users_df
```

```python
users_df.dropna(subset=['Country'], inplace=True)
users_df
```

```python
users_new_df=pd.merge(ratings_df, users_df, on='User-ID')
users_new_df.shape
```

```python
# Delete extraneous column
del users_new_df['Book-Rating']
users_new_df.head()
```
